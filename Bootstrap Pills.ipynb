```python
from IPython.display import HTML, display

badge_class = (
    "bg-success-subtle text-success-emphasis rounded-pill"
    if is_success
    else "bg-danger-subtle text-danger-emphasis rounded-pill"
)

label = "Success" if is_success else "Danger"

display(HTML(f"""
<style>{css}</style>

Test: 
<span class="badge {badge_class}">
  {label}
</span>
"""))
```

```python
def status_badge(ok: bool, prefix: str = ""):
    cls = (
        "bg-success-subtle text-success-emphasis rounded-pill"
        if ok
        else "bg-danger-subtle text-danger-emphasis rounded-pill"
    )
    label = "Success" if ok else "Fail"

    return f"""
    <span style="margin-right: 0.5rem;">{prefix}:</span>
    <span class="badge {cls}">{label}</span>
    """
```

```python
display(HTML(status_badge(Validation1, "Text")))
```
